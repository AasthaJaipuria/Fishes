```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_pickle("clean_data2.pkl")
```

```python
data = data.loc[:,['ID','Species Name','Env Section']]
```

```python
data['refCount'] = data['Env Section'].apply(lambda x : x.count('Ref.'))
```

```python
data['refList'] = data['Env Section'].apply(lambda x : re.findall('depth range [\?0-9]+ \- [\?0-9]+ m.\(Ref..\d+|\S*.\(Ref..\d+',x))
```

```python
data['mig_ref'] = ''
data['climate_ref'] = ''
data['depth_ref'] = ''
data['last_ref'] = ''
data['habitat_ref'] = ''
```

```python
migrations = ['anadromous', 'catadromous', 'amphidromous', 'potamodromous', 'limnodromous', 'oceanodromous', 'non-migratory']
habitats = ['benthopelagic', 'pelagic', 'demersal', 'pelagic-neritic', 'reef-associated', 'bathypelagic', 
            'bathydemersal', 'pelagic-oceanic']
for ID in data.index:
    if data['refCount'][ID]==0:
        continue
    for ref in data['refList'][ID]:
        if ref.startswith(tuple(migrations)):
            data['mig_ref'][ID] = ref.split('Ref. ')[1]
            break
    for ref in data['refList'][ID]:
        if ref.startswith('depth'):
            data['depth_ref'][ID] = ref.split('Ref. ')[1]
            break
    for ref in data['refList'][ID]:
        if ref.startswith('Tropical'):
            data['climate_ref'][ID] = ref.split('Ref. ')[1]
            break
    for ref in data['refList'][ID]:
        if ref.startswith(tuple(habitats)):
            data['habitat_ref'][ID] = ref.split('Ref. ')[1]
            break
    for ref in data['refList'][ID]:
        if '°' in ref:
            data['last_ref'][ID] = ref.split('Ref. ')[1]
            break
        elif '?' in ref and not ref.startswith('depth'):
            data['last_ref'][ID] = ref.split('Ref. ')[1]
            break
```

```python
data.iloc[:,-5:] = data.iloc[:,-5:].replace('',np.nan)
```

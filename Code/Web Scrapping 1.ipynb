```python
import requests
from bs4 import BeautifulSoup
import string
```

```python
fishes = []
for x in string.ascii_uppercase:
    html_text = requests.get("https://www.fishbase.de/ListByLetter/ScientificNames{}.htm".format(x)).text
    soup = BeautifulSoup(html_text,'html.parser')
    count = 0
    for link in soup.find_all('a'):
        if len(link.text.split())==2:
            count +=1
            fishes.extend([[int(str(link.get('href')).split('=')[1]),link.text]])
    print(x,count)
```

```python
#Total Species
len(fishes)
```
